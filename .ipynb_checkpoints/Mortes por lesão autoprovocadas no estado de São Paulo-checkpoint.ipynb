```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
low_memory=False,
%matplotlib inline
%config InlineBackend.figure_format='svg'
pd.options.display.max_columns = 80
pd.options.display.max_rows = 90
```

```python
filename = 'C:/Users/Rafael Rezende/Desktop/Analise_de_dados_obito/obitos_por_suicidio_regiao_metropolitana.xlsx'
df = pd.read_excel(filename)
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams['xtick.labelsize'] = 8
plt.pie(df['Branca'],labels = df['Região Metropolitana 0 RIDE'],
       autopct='%1.1f%%',
       pctdistance=1.045, labeldistance=1.106)
plt.title('Regiões do estado de São Paulo onde, em 2018, ocorreram mortes\n por lesões autoprovocadas de pessoas de cor/raça branca', loc="center")
plt.legend(df['Região Metropolitana 0 RIDE'], title="Regiões Metropolitanas", loc="upper right", bbox_to_anchor=(0.593,0,0.52,1.05))
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams['xtick.labelsize'] = 8
plt.pie(df['Preta'],labels = df['Região Metropolitana 0 RIDE'],
       autopct='%1.1f%%',
       pctdistance=1.05, labeldistance=1.116)
plt.title('Regiões do estado de São Paulo onde, em 2018, ocorreram mortes\n por lesões autoprovocadas de pessoas de cor/raça preta', loc="center")
plt.legend(df['Região Metropolitana 0 RIDE'], title="Regiões Metropolitanas", loc="upper right", bbox_to_anchor=(0.5377,0,0.52,1.05))
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams['xtick.labelsize'] = 8
plt.pie(df['Amarela'],labels = df['Região Metropolitana 0 RIDE'],
       autopct='%1.1f%%',
       pctdistance=1.045, labeldistance=1.106)
plt.title('Regiões do estado de São Paulo onde, em 2018, ocorreram mortes\n por lesões autoprovocadas de pessoas de cor/raça amarela', loc="center")
plt.legend(df['Região Metropolitana 0 RIDE'], title="Regiões Metropolitanas", loc="upper right", bbox_to_anchor=(0.544,0,0.52,1.05))
plt.show()
```

## Com base nos gráficos acima dá pra tirar de conclusão que menos de 7% das mortes de brancos e negros por lesões autoprovocadas ocorreram no Vale do Paraíba e Litoral Norte, se tornando assim a quarta região com menos casos

## E em relação a pessoas de raça amarela essa porcentagem cai pra 4,5% de mortes na região do Vale do Paraíba e Litoral Norte, fazendo com que assim a mesma seja a quinta região com menos casos
